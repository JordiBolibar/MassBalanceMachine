```python
import pandas as pd
```

```python
df = pd.read_csv('../data/files/Iceland_Stake_Data_Reprojected.csv')
```

```python
df
```

```python
melted_df = pd.melt(df, id_vars=['stake','ba_stratigraphic', 'bw_stratigraphic', 'bs_stratigraphic'], value_vars=['d1', 'd2', 'd3'], var_name='Measurement Type', value_name='Measurement Date', ignore_index=True)

# Create separate DataFrames for each SMB type and corresponding date
winter_df = melted_df[melted_df['Measurement Type'] == 'd1'].copy()
winter_df['SMB'] = winter_df['bw_stratigraphic']
winter_df['Type'] = 'BW'

summer_df = melted_df[melted_df['Measurement Type'] == 'd2'].copy()
summer_df['SMB'] = summer_df['bs_stratigraphic']
summer_df['Type'] = 'BS'

annual_df = melted_df[melted_df['Measurement Type'] == 'd3'].copy()
annual_df['SMB'] = annual_df['ba_stratigraphic']
annual_df['Type'] = 'BA'

# Select only necessary columns
winter_df = winter_df[['Stake', 'Measurement Date', 'SMB', 'Type']]
summer_df = summer_df[['Stake', 'Measurement Date', 'SMB', 'Type']]
annual_df = annual_df[['Stake', 'Measurement Date', 'SMB', 'Type']]

# Concatenate the DataFrames
result_df = pd.concat([winter_df, summer_df, annual_df]).reset_index(drop=True)
```
